# 4.Finding the scoring variation!?

* Scoring variation in case of gender:

```python
#average of math,reading,writing score in case of gender
data.groupby("gender")[['math score','reading score','writing score']].mean().round(1)
```

```python
#minimum in case of gender
data.groupby("gender")[['math score','reading score','writing score']].min()
```

```python
#maximum in case of gender
data.groupby("gender")[['math score','reading score','writing score']].max()
```

Insight:

1.Analyzing the data for average values of math,reading,writing score the consistent score is maintainted by male.

2.Analyzing the data in case of minimum values for math,reading,writing the minimum values are gained by females, ie the better ones are males.

3.Analyzing the data in case of maximum values for math,reading,writing the maximum values are 100 in case of both female and male.

 * Scoring variation in case of 'Test preparation course'

```python
#average value
data.groupby("test preparation course")[['math score','reading score','writing score']].mean().round(1)
```

```python
#minimum value
data.groupby("test preparation course")[['math score','reading score','writing score']].median()
```

```python
#maximum value
data.groupby("test preparation course")[['math score','reading score','writing score']].max()
```

Insight:

1.After analyzing the data the one who completed test preparation course has less variation compared to none.
